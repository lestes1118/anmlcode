```python
import pandas as pd #ensure to download software. pandas used to deal with xlx files 
import matplotlib.pyplot as plt
```

```python
#insure to move your desired file into jupyter lab folder/terminal where code will be ran 
file='Secondwashingstep.xlsx' #change file name here
data=pd.read_excel(file)
(data.head())
#loaded correctly
```

```python

x=data['Subtracted_Data X1'][2:7502]  #put name of top row and the row numbers where in python 1 starts at 0.
y=data['Subtracted_Data Y1'][2:7502] 

plt.figure(figsize=(8,6))
plt.plot(x,y,linestyle='-',color='r')
plt.xlabel('2 theta [degs.]',fontsize=15,fontname='Arial')
plt.ylabel('Intensity [counts]',fontsize=15,fontname='Arial')
plt.title('XRD Data', fontsize=20, fontname='Arial')
plt.show()

```
